```python
# Create a table

conn = psycopg2.connect("dbname=deme_db user=postgres password=***")
cur = conn.cursor()
command = '''create table urun
(
    id integer primary key,
    ad varchar(15) NOT NULL,
    marka varchar(20) NOT NULL,
    stok integer,
    kategori varchar(15)
)'''
#create a table
cur.execute(command)
# close communication with the PostgreSQL database server
cur.close()
# commit the changes
conn.commit()
# close the connection
conn.close()
```

```python
# Insert data

conn = psycopg2.connect("dbname=deme_db user=postgres password=***")
cur = conn.cursor()
cur.execute('INSERT INTO musteri VALUES(%s,%s,%s,%s,%s,%s)', (13,'asim','tan','izmir', 4000, 5))
cur.close()
conn.commit()
conn.close()
```

```python
# update data

conn = psycopg2.connect("dbname=deme_db user=postgres password=***")
cur = conn.cursor()
cur.execute('UPDATE musteri SET meslek=%s WHERE id=%s', (4,3))
cur.close()
conn.commit()
conn.close()
```

```python
# querying data

conn = psycopg2.connect("dbname=deme_db user=postgres password=***")
cur = conn.cursor()
cur.execute("SELECT * FROM urun")
rows = cur.fetchall()
for row in rows:
    print(row)
cur.close()
conn.close()
```
